4. Write a Python program to read last n lines of a file.

```python
def last_line(filename, n):
    with open(filename, 'r') as f:
        line = f.readlines()
        print(line)

        for l in line[-n:]:
            print(l.strip())

last_line('..\sample.txt', 2)
```

5. Write a Python program to read a file line by line and store it into a list.

```python
def read_line(file_name):
    lst = []
    with open(file_name , 'r')as f:
        for l in f:
            lst.append(l.strip())
    return lst

read_line('..\sample.txt')
```

6. Write a Python program to read a file line by line store it into a variable.

```python
def store_in_var(filename):
    variable = ''
    with open(filename, 'r') as f:
        for l in f:
            variable +=l
        return variable
store_in_var('..\sample.txt')
```

9. Write a Python program to count the frequency of words in a file.


```python
import string

def word_frequency(file_name):
    freq = {}
    
    with open(file_name, 'r') as file:
        for line in file:
            words = line.split()
            
            for word in words:
                # normalize word
                word = word.lower()
                
                if word:
                    freq[word] = freq.get(word, 0) + 1
    
    return freq

# Example usage
result = word_frequency('..\sample.txt')
print(result)
```

12. Write a Python program to copy the contents of a file to another file .


```python
def copy_file(source, destination):
    with open(source, 'r') as src, open(destination, 'w') as dest:
        content = src.read()
        dest.write(content)

# Example usage
copy_file('..\sample.txt', 'sample2.txt')
```

13. Write a Python program to read a random line from a file.


```python
import random

def random_line(file_name):
    with open(file_name, 'r') as file:
        lines = file.readlines()
    
    return random.choice(lines).strip()

# Example usage
print(random_line('sample2.txt'))
```

14. Write a Python program to assess if a file is closed or not.


```python
def check_file_closed(file_name):
    file = open(file_name, 'r')
    
    print("Before closing:", file.closed)  # False
    
    file.close()
    
    print("After closing:", file.closed)   # True

# Example usage
check_file_closed('sample2.txt')
```

15.Write a Python program that takes a text file as input and returns the number of words of a given text file.

```python
def count_words(file_name):
    with open(file_name, 'r') as file:
        text = file.read()
    
    words = text.split()
    return len(words)

# Example usage
print(count_words('sample2.txt'))
```
