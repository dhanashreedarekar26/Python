6. Write a Python program to create a class representing a stack data structure. Include methods for pushing and popping elements.


```python
class Stack:
    
    def __init__(self):
        self.data = []
    
    def  push(self,value):
        self.data.append(value)

    def pop(self):
        if self.is_empty():
            raise IndexError('stack is empty: cannot pop from the empty stack!!')
        return self.data.pop()
    
    def is_empty(self):
        return len(self.data) == 0    
    
    def display_stack(self):
        print(self.data)
    

stack = Stack()
stack.push(20)
stack.push(100)
stack.push(200)
stack.push(30)
stack.pop()
stack.display_stack()

```

```python
class Shopping_cart:

    def __init__(self):
        self.items = {}

    def add(self, name, price, qte = 1):
        if name in self.items:
            self.items[name]['qte']+=qte
        else:
            self.items[name] = {'price': price, 'qte': qte}

    def remove(self, name):
        if name in self.items:
            del self.items[name]
        else:
            print('Item not found in cart')

    def cal_total(self):
        total = 0
        for item in self.items.values():
            total += item['price'] * item['qte']
        print(f'Total of the cart is: {total}')
    
    def display(self):
        if not self.items:
            print('cart is empty')
        else:
            for name, details in self.items.items():
                print(f'{name} ->Price: {details['price']}, quantity:{ details['qte']}')

cart = Shopping_cart()
cart.add('25 ultra', 130000)
cart.add('24 ultra', 10000)
cart.add('23 ultra', 60000)
cart.add('25 ultra', 130000)
cart.remove('25 ultra')
cart.cal_total()
cart.display()

```

```python
class BankAccount:
    
    # Constructor
    def __init__(self, accountNumber, name, balance):
        self.accountNumber = accountNumber
        self.name = name
        self.balance = balance

    
    def Deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited: {amount}")
        else:
            print("Invalid deposit amount!")


    def Withdrawal(self, amount):
        if amount <= 0:
            print("Invalid withdrawal amount!")
        elif amount > self.balance:
            print("Insufficient balance!")
        else:
            self.balance -= amount
            print(f"Withdrawn: {amount}")

    # Bank fees method (5% deduction)
    def bankFees(self):
        fee = self.balance * 0.05
        self.balance -= fee
        print(f"Bank fees applied: {fee}")

    def display(self):
        print("\nAccount Details:")
        print(f"Account Number : {self.accountNumber}")
        print(f"Account Holder : {self.name}")
        print(f"Balance        : {self.balance}")


acc1 = BankAccount(101, "Dhanashree", 1000)

acc1.Deposit(500)
acc1.Withdrawal(300)
acc1.bankFees()

acc1.display()



```

	Example 1:

	Approach:

	Create a class named FlashCard.
	Initialize dictionary fruits using init() method. Here you have to define fruit name as key and it's color as value. E.g., {"Banana": "yellow", "Strawberries": "pink"}
	Now randomly choose a pair from fruits by using random module and store the key in variable fruit and value in variable color.
	Now prompt the user to answer the color of the randomly chosen fruit.

	welcome to fruit quiz

	pink

```python
import random
class Flashcard:

    def __init__(self):
        self.fruits = {"Banana": "yellow",
            "Strawberries": "pink",
            "Watermelon": "green",
            "Apple": "red",
            "Mango": "yellow"}
    
    def play(self):
        print('Lets start the Games!!!')
        while True:
            name , color = random.choice(list(self.fruits.items()))
            

            answer = input(f'Answer the color of {name}')
            if answer.lower() == color.lower():
                print('you are correct!!')
        
            else:
                print('wrong answer!!')

            choice = input("Enter 0, if you want to play again: ")
            if choice != '0':
                break
    
flash = Flashcard()
flash.play()

    
```

	overtime = hours_worked - 50
	Overtime amount = (overtime * (salary / 50))
	

```python
class Employee:
    
    def __init__(self):
        self.info = {}
    
    def add_emp(self,emp_id, emp_name, emp_sal, emp_depart):
        self.info[emp_id]={
            'emp_name' : emp_name,
            'emp_salary': emp_sal,
            'emp_department' : emp_depart 
        }
        return self.info
    
    def assign_department(self, emp_id, emp_department):
        if emp_id not in self.info:
            print('Employee Id is invalid!!')
        else:
            self.info[emp_id]['emp_department'] = emp_department
        print(self.info)

    def print_details(self, emp_id):
        if emp_id not in self.info:
            print('This id is not valid!!')
        else:
            data = self.info[emp_id]
            print(f'Employee Name: {data['emp_name']}')
            print(f'Employee Id: {data['emp_name']}')
            print(f'Employee Salary: {data['emp_salary']}')
            print(f'Employee Department: {data['emp_department']}')

    def calculate_emp_salary(self, emp_sal, hours_worked):
        if emp_sal>50:
            overtime = hours_worked - 50
            total_amt = overtime * (emp_sal / 50)
            emp_sal += total_amt
            return 'Salary: ', str(emp_sal)
        
            

emp = Employee()
emp.add_emp(11, 'Dhanashree Darekar', 200000, 'IT')
emp.assign_department(11, 'Marketing')
emp.print_details(11)
emp.calculate_emp_salary(20000, 90)

```

19. Write a Python class Restaurant with attributes like menu_items, book_table, and customer_orders, and methods like add_item_to_menu, book_tables, and customer_order.

	Perform the following tasks now:

	Now add items to the menu.
	Make table reservations.
	Take customer orders.
	Print the menu.
	Print table reservations.
	Print customer orders.
	Note: Use dictionaries and lists to store the data.

```python
class Restaurant:
    def __init__(self):
        self.menu_items = {}
        self.book_tables = []
        self.customer_orders ={}

    def add_item_to_menu(self,dish, price):
        self.menu_items[dish]={'Price': price}
        print(f'{self.menu_items[dish]} Is Added to Menu succesfully!!!!')
    
    def book_table(self,table_num):
        if table_num in self.book_tables:
            print('Table is already Booked')
        else:
            self.book_tables.append(table_num)
            print('Table booked Succesfully!!')

    def customer_order(self, table_number, order_items):
        if table_number not in self.book_tables:
            print("Table not reserved!")
        else:
            self.customer_orders[table_number] = order_items
            print(f"Order for table {table_number} recorded.")

    def print_menu(self):
        print("\n--- MENU ---")
        for item, price in self.menu_items.items():
            print(f"{item} : ₹{price}")


    def print_table_reservations(self):
        print("\n--- TABLE RESERVATIONS ---")
        for table in self.book_tables:
            print(f"Table {table}")


    def print_customer_orders(self):
        print("\n--- CUSTOMER ORDERS ---")
        for table, orders in self.customer_orders.items():
            print(f"Table {table}: {', '.join(orders)}")


r = Restaurant()
r.add_item_to_menu("Pizza", 250)
r.add_item_to_menu("Burger", 120)
r.add_item_to_menu("Pasta", 200)

r.book_table(1)
r.book_table(2)

r.customer_order(1, ["Pizza", "Pasta"])
r.customer_order(2, ["Burger"])

r.print_menu()
r.print_table_reservations()
r.print_customer_orders()
```
